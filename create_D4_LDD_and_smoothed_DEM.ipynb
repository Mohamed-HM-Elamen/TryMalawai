This notebook creates a D4 LDD and a smoothed DEM based on this LDD.

The D4 LDD is required since the smoothed DEM will be applied to a 3Di subgrid model which cannot handle diagonal directions, and the DEM should be smoothed in a way that reflects the potential flow pattern within the model. The 'normal' D8 LDD (as present in PCRaster functions) is therefore not suitable and a new script was required to create a D4 LDD. This is based on a pre-processed DEM which was burned and only includes values within the relevant streamlines (with no data values everywhere else), to make the creation of such a LDD much simpler/faster (and the smoothing is only required within the streamlines anyway).

The smoothing algorithm is carried out after the creation of the LDD and uses the shapefile with streamlines (which was used to extract the relevant grid cells for the pre-processed grid) to identify the most upstream locations as starting points for the smoothing. From these starting points the algorithm checks all grid cells from each starting point to either the outflow point or until a cell is reached that has been checked by the algorithm before (based on a user-defined option).

Attempting to create a LDD, starting from the outflow point and moving upstream:

```python
# create a LDD map with NoData at the same locations of NoData as the DEM, but with 0 values everywhere else
LDD_np = np.copy(DEM_np)

for i in range(len(DEM_np)):
    for j in range(len(DEM_np[0])):
        if DEM_np[i][j] != -999:
            LDD_np[i][j] = 0
```

```python
# define LDD values
LDD_value_y_plus1 = 8
LDD_value_y_min1  = 2
LDD_value_x_plus1 = 4
LDD_value_x_min1  = 6
LDD_value_pit     = 5

LDD_value_list = [LDD_value_y_plus1, LDD_value_y_min1, LDD_value_x_plus1, LDD_value_x_min1]

# assign a pit at the outflow point
LDD_np[outflow_index_double] = LDD_value_pit

# current location (double indices)
current_location = outflow_index_double

# list of remaining locations
list_remaining_locations = []

# count
count = 0
break_count = 20000    # used to break while-loop (to prevent a potential infinite loop)

# loop over array until all zero values have been changed
while 0 in LDD_np:
    
    # count
    count += 1
    
    # possible directions
    location_y_plus1 = (current_location[0]+1,current_location[1])
    if current_location[0]+1 >= len(DEM_np):
        location_y_plus1 = -999.
        
    location_y_min1  = (current_location[0]-1,current_location[1])
    if current_location[0]-1 < 0:
        location_y_min1 = -999.
        
    location_x_plus1 = (current_location[0],current_location[1]+1)
    if current_location[1]+1 >= len(DEM_np[0]):
        location_x_plus1 = -999.
        
    location_x_min1  = (current_location[0],current_location[1]-1)
    if current_location[1]-1 < 0:
        location_x_min1 = -999.
    
    location_list = [location_y_plus1, location_y_min1, location_x_plus1, location_x_min1]
    
    # elevation values at possible connected cells
    if location_y_plus1 != -999:
        DEM_y_plus1 = DEM_np[location_y_plus1]
    else:
        DEM_y_plus1 = -999.
    if location_y_min1 != -999:
        DEM_y_min1  = DEM_np[location_y_min1]
    else:
        DEM_y_min1 = -999.
    if location_x_plus1 != -999:
        DEM_x_plus1 = DEM_np[location_x_plus1]
    else:
        DEM_x_plus1 = -999.
    if location_x_min1 != -999:
        DEM_x_min1  = DEM_np[location_x_min1]
    else:
        DEM_x_min1 = -999.
    
    # LDD values at possible connected cells
    if location_y_plus1 != -999:
        LDD_y_plus1 = LDD_np[location_y_plus1]
    else:
        LDD_y_plus1 = -999.
    if location_y_min1 != -999:
        LDD_y_min1  = LDD_np[location_y_min1]
    else:
        LDD_y_min1 = -999.
    if location_x_plus1 != -999:
        LDD_x_plus1 = LDD_np[location_x_plus1]
    else:
        LDD_x_plus1 = -999.
    if location_x_min1 != -999:
        LDD_x_min1  = LDD_np[location_x_min1]
    else:
        LDD_x_min1 = -999.
    
    list_DEM = [DEM_y_plus1, DEM_y_min1, DEM_x_plus1, DEM_x_min1]
    list_LDD = np.array([LDD_y_plus1, LDD_y_min1, LDD_x_plus1, LDD_x_min1])
    
    # calculate elevation difference with current cell (for those with LDD = 0)
    elevation_difference = np.ones(4) * -999
    for i in range(len(list_DEM)):
        if (list_DEM[i] != -999) and (list_LDD[i] == 0):
            elevation_difference[i] = list_DEM[i] - DEM_np[current_location]
    
    # find upstream direction (maximum elevation difference) and store other potential directions
    try:
        
        if len(np.where(elevation_difference==-999)[0]) != 4:
            upstream_direction = np.where(elevation_difference==max(elevation_difference))[0][0]
            
            if len(np.where(elevation_difference!=-999)[0]) > 1:
                list_remaining_locations.append(current_location)
                
        else:
            upstream_direction = None
                
    except:
        upstream_direction = None
    
    if upstream_direction != None:
        
        # update current location
        current_location = location_list[upstream_direction]
        
        # adjust LDD map
        LDD_np[current_location] = LDD_value_list[upstream_direction]
        
    else:
        
        # check if there are any remaining locations
        if len(list_remaining_locations) > 0:
            current_location = list_remaining_locations[0]
            del list_remaining_locations[0]
            
        # if not, stop algorithm (all points should be looped over)
        else:
            print 'could not find any more points!'
            print 'algorithm stopped!'
            break
    
    # to break potential infinite loop
    if count >= break_count:
        break
```

Now everything is in place for the actual smoothing:

```python
DEM_np_smoothed = np.copy(DEM_np)
DEM_np_burned   = np.copy(DEM_np)
```

```python
# lists with original and adjusted DEM values (to inspect/plot)
DEM_list_original      = [[None]] * len(upstream_indices_single)
DEM_list_smoothed      = [[None]] * len(upstream_indices_single)
DEM_list_burned        = [[None]] * len(upstream_indices_single)

# list with locations looped over during smoothing and burning
list_locations_loop = [[None]] * len(upstream_indices_single)

# count iterations per upstream location
count = np.zeros(len(upstream_indices_single))
```

```python
# loop over all specified upstream points
for k in range(len(upstream_indices_single)):

    # value to control while loop
    found_pit = 0

    # starting location
    last_location  = upstream_indices_double[k]

    # write values at starting location to lists
    DEM_list_original[k]   = [DEM_np[last_location]]
    DEM_list_smoothed[k]   = [DEM_np_smoothed[last_location]]
    DEM_list_burned[k]     = [DEM_np_burned[last_location]]

    list_locations_loop[k] = [last_location]

    # loop from upstream starting location to most downstream point
    while found_pit == 0:
        
        # direction at last known location
        temp_direction = LDD_np[last_location]
        
        if temp_direction == 5:
            found_pit = 1
            break
        elif temp_direction == 6:
            new_location = (last_location[0] + 0, last_location[1] + 1)
        elif temp_direction == 2:
            new_location = (last_location[0] + 1, last_location[1] + 0)
        elif temp_direction == 4:
            new_location = (last_location[0] + 0, last_location[1] - 1)
        elif temp_direction == 8:
            new_location = (last_location[0] - 1, last_location[1] + 0)

            
        # check if while-loop should stop here because value was already looped over before
        if k > 0:
            if stop_same_point == 1:
                for i in range(len(list_locations_loop)):
                    if (new_location in list_locations_loop[i]) and (i != k):
                        # write this final location to list of locations
                        list_locations_loop[k].append(new_location)
                        # stop while loop
                        found_pit = 1
                        break
                        
        # write new location to list of locations
        list_locations_loop[k].append(new_location)

        # compare DEM value at new location with last location
        temp_difference_DEM = DEM_np[new_location] - DEM_np_smoothed[last_location]

        # adjust value if DEM new location > DEM last location
        if temp_difference_DEM > 0.:
            DEM_np_smoothed[new_location] = DEM_np_smoothed[last_location]
            
        # apply stream burning
        DEM_np_burned[new_location] = DEM_np_smoothed[new_location]
        DEM_np_burned[new_location] -= apply_burning

        # add values to lists
        DEM_list_original[k].append(DEM_np[new_location])
        DEM_list_smoothed[k].append(DEM_np_smoothed[new_location])
        DEM_list_burned[k].append(DEM_np_burned[new_location])

        # update location
        last_location = new_location

        # count iterations
        count[k] += 1
```
