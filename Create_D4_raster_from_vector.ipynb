This notebook explores the possiblities of creating a D4 raster representation (map) of a vector streamline (shapefile). This is useful when attempting to condition a DEM based on a vector streamline (e.g. derived from satellite imagery), if the DEM is to be used in a model that does not allow diagional movement between cells (such as 3Di and Delft3D FM). In GIS the vector can easily be converted to a raster, but this results in sections with diagonal 'movement'.

```python
# convert to separate x and y coords lists
list_points_x = []
list_points_y = []
for i in range(len(shapes)):
    temp_list_x = []
    temp_list_y = []
    for j in range(len(shapes[i].points)):
        temp_list_x.append(shapes[i].points[j][0])
        temp_list_y.append(shapes[i].points[j][1])
    list_points_x.append(temp_list_x)
    list_points_y.append(temp_list_y)
```

```python
# plot
plt.figure(figsize=(20,8))
for i in range(len(list_points_x)):
    plt.plot(list_points_x[i], list_points_y[i],'o')
plt.autoscale()
plt.show()
```

```python
# get coords of DEM
coords_DEM = pcr_spatial_functions.getAllCellCoords(pcr.boolean(DEM_map/DEM_map))
```

What would be the easiest way to find all grid cells through which the streamline passes? Just using the points won't work, as these might skip a grid cell and also won't be able to use non-diagonal directions...Perhaps the easiest approach would be to use the Shapely library, convert the streamline to Lines and DEM grid to Polygons and then check intersections?

```python
# create list of Polygons
list_DEM_Polygons = []
for i in range(len(coords_DEM)):
    list_DEM_Polygons.append(Polygon(coords_DEM[i]))
```

```python
#DEM_MultiPolygon = MultiPolygon(list_DEM_Polygons)
```

```python
# create list of LineStrings
list_lines = []
for i in range(len(shapes)):
    list_lines.append(LineString(shapes[i].points))
```

```python
# create one MultiLineString
Streamline_MultiLineString = MultiLineString(list_lines)
```

```python
# check all polygons against intersection with the MultiLineString
list_intersection_ids = []
for i in range(len(list_DEM_Polygons)):
    if list_DEM_Polygons[i].intersects(Streamline_MultiLineString):
        list_intersection_ids.append(i)

print 'number of intersected polygons found:', len(list_intersection_ids)
```

```python
# the above result is in single indices, double indices are required to acces numpy arrays
list_intersection_ids_double = pcr_spatial_functions.convertSingleMapIndicesToDouble(list_intersection_ids, \
                                                                                     pcr.boolean(DEM_map/DEM_map), \
                                                                                     255)
```

```python
# create numpy arrays with zero's (to update with indices calculated above)
zero_map_np = pcr.pcr2numpy((DEM_map/DEM_map) - 1, -999)

zero_map_np_masked = np.ma.masked_equal(zero_map_np, -999)

# plot to check
plt.figure(figsize=(20,8))
plt.imshow(zero_map_np_masked, vmin=-999, vmax=0)
plt.colorbar()
plt.autoscale()
plt.show()
```

```python
# create 0/1 map where 1 indicates location of stream
stream_raster = np.copy(zero_map_np_masked)

for i in range(len(list_intersection_ids_double)):
    stream_raster[list_intersection_ids_double[i]] = 1.
    
stream_raster_masked = np.ma.masked_equal(stream_raster, -999)

# plot to check
plt.figure(figsize=(20,8))
plt.imshow(stream_raster_masked, cmap='brg_r', vmin=0., vmax=1.)
plt.colorbar()
plt.autoscale()
plt.show()
```

```python
# convert numpy array to map to save to file
stream_map = pcr.numpy2pcr(pcr.Scalar, stream_raster, -999)

path_output = r"D:\haag\Files\Projects\RASOR\Casestudy_Malawi\data\Other\StreamLocations_D4_Salima_3.map"

#pcr.report(stream_map, path_output)
```

Also create a version with a buffer around the streamline, to prevent very small areas of high refinement in FM.

```python
buffer_window = 1

stream_raster_buffer = np.copy(zero_map_np_masked)

for i in range(len(list_intersection_ids_double)):
    temp_cut_area = general_spatial_functions.cut_area(stream_raster_buffer, list_intersection_ids_double[i], buffer_window)
    for p in range(len(temp_cut_area[1])):
        for q in range(len(temp_cut_area[2])):
            stream_raster_buffer[temp_cut_area[2][q]][temp_cut_area[1][p]] = 1.

stream_raster_buffer_masked = np.ma.masked_equal(stream_raster_buffer, -999)

# plot to check
plt.figure(figsize=(20,8))
plt.imshow(stream_raster_buffer_masked, cmap='brg_r', vmin=0., vmax=1.)
plt.colorbar()
plt.autoscale()
plt.show()
```

```python
# convert numpy array to map to save to file
stream_buffer_map = pcr.numpy2pcr(pcr.Scalar, stream_raster_buffer, -999)

path_output2 = r"D:\haag\Files\Projects\RASOR\Casestudy_Malawi\data\Other\StreamLocations_D4_Salima_3_buffer1.map"

#pcr.report(stream_buffer_map, path_output2)
```

Add an additional buffer to the 'main' stream, i.e. the section after the confluence:

```python
main_stream_ids = [0,3,4,5]

# plot
plt.figure(figsize=(20,8))
for i in range(len(list_points_x)):
    if i in main_stream_ids:
        plt.plot(list_points_x[i], list_points_y[i],'-b', linewidth=2)
    else:
        plt.plot(list_points_x[i], list_points_y[i],'--b')
plt.autoscale()
plt.show()
```

```python
# create list of LineStrings and one MultiLineString for these only
list_lines_main =  []
list_lines_other = []
for i in range(len(shapes)):
    if i in main_stream_ids:
        list_lines_main.append(LineString(shapes[i].points))
    else:
        list_lines_other.append(LineString(shapes[i].points))

StreamlineMain_MultiLineString  = MultiLineString(list_lines_main)
StreamlineOther_MultiLineString = MultiLineString(list_lines_other)

# check all polygons against intersection with this MultiLineString
list_intersection_ids_main  = []
list_intersection_ids_other = []
for i in range(len(list_DEM_Polygons)):
    if list_DEM_Polygons[i].intersects(StreamlineMain_MultiLineString):
        list_intersection_ids_main.append(i)
    elif list_DEM_Polygons[i].intersects(StreamlineOther_MultiLineString):
        list_intersection_ids_other.append(i)

print 'number of intersected polygons found (main): ', len(list_intersection_ids_main)
print 'number of intersected polygons found (other):', len(list_intersection_ids_other)
```

```python
# the above result is in single indices, double indices are required to acces numpy arrays
list_intersection_ids_main_double = pcr_spatial_functions.convertSingleMapIndicesToDouble(list_intersection_ids_main, \
                                                                                          pcr.boolean(DEM_map/DEM_map), \
                                                                                          255)

list_intersection_ids_other_double = pcr_spatial_functions.convertSingleMapIndicesToDouble(list_intersection_ids_other, \
                                                                                           pcr.boolean(DEM_map/DEM_map), \
                                                                                           255)
```

```python
buffer_window_main  = 2
buffer_window_other = 1

stream_raster_buffer_main = np.copy(zero_map_np_masked)

for i in range(len(list_intersection_ids_main_double)):
    temp_cut_area = general_spatial_functions.cut_area(stream_raster_buffer_main, list_intersection_ids_main_double[i], buffer_window_main)
    for p in range(len(temp_cut_area[1])):
        for q in range(len(temp_cut_area[2])):
            stream_raster_buffer_main[temp_cut_area[2][q]][temp_cut_area[1][p]] = 1.

for i in range(len(list_intersection_ids_other_double)):
    temp_cut_area = general_spatial_functions.cut_area(stream_raster_buffer_main, list_intersection_ids_other_double[i], buffer_window_other)
    for p in range(len(temp_cut_area[1])):
        for q in range(len(temp_cut_area[2])):
            stream_raster_buffer_main[temp_cut_area[2][q]][temp_cut_area[1][p]] = 1.

stream_raster_buffer_main_masked = np.ma.masked_equal(stream_raster_buffer_main, -999)

# plot to check
plt.figure(figsize=(20,8))
plt.imshow(stream_raster_buffer_main_masked, cmap='brg_r', vmin=0., vmax=1.)
plt.colorbar()
plt.autoscale()
plt.show()
```

```python
# convert numpy array to map to save to file
stream_buffer_main_map = pcr.numpy2pcr(pcr.Scalar, stream_raster_buffer_main, -999)

path_output5 = r"D:\haag\Files\Projects\RASOR\Casestudy_Malawi\data\Other\StreamLocations_D4_Salima_3_buffer_2_1.map"

#pcr.report(stream_buffer_main_map, path_output5)
```

Since the file is to be used in FM, which only takes -999 as missing data and requires it to be converted to .xyz file, it might be good to already set these missing values correctly:

```python
# first convert to numpy, setting missing values to -999
dem_burn_stream_np = pcr.pcr2numpy(dem_burn_stream_map, -999)

# then convert back to a PCRaster map, by not specifying the correct missing values (so that -999 values stay the same)
dem_burn_stream_map_NoData999 = pcr.numpy2pcr(pcr.Scalar, dem_burn_stream_np, -1)

# output file
path_output4 = os.path.join(r"D:\haag\Files\Projects\RASOR\Casestudy_Malawi\data\DEM\SRTM_30m", 'SRTM_30m_Salima_adj_buffer_2_1_burn_' + \
                                                                                                str(burn_buffer_value) + '_' + \
                                                                                                str(burn_stream_value) + \
                                                                                                '_NoData999.map')

pcr.report(dem_burn_stream_map_NoData999, path_output4)
```
