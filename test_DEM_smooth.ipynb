Now the actual algorithm:

```python
# smoothing/burning/filling algorithm, with burning next to main channel

# it first runs the smoothing algorithm, and then does the burning in a next loop

# ---------------------------------------------------------------------------------------------
# apply burning as well as smoothing?
apply_burning_main = 3.   # burning of main channel (LDD path) [m]
apply_burning_wide = 2.   # burning next to main channel [m]
limit_to_zero = 1         # 0 = no, 1 = yes

# stop when encountering a location that was already changed in a previous loop
# (e.g. if second loop for Bani comes to a point already changed for the Niger)
# 0 = no, 1 = yes
stop_same_point = 1

# threshold value to identify possible outliers
outlier_threshold = 2.

# number of points to consider to assess possible outliers
outlier_points = 5
# ---------------------------------------------------------------------------------------------

# lists with original and adjusted DEM values (to inspect/plot)
DEM_list_original      = [[None]] * len(list_upstream_point_locs)
DEM_list_adjusted      = [[None]] * len(list_upstream_point_locs)
DEM_list_burned_main   = [[None]] * len(list_upstream_point_locs)

# list with locations looped over during smoothing and burning
list_locations_loop = [[None]] * len(list_upstream_point_locs)
list_locations_loop_burn_wide = []

# count iterations per upstream location
count = np.zeros(len(list_upstream_point_locs))

# initialze 2 loops (1 for smoothing and main burning, 1 for wide burning)
for p in range(2):
    
    # copy smoothed and mainstream burned DEM to DEM for wider burning
    if p == 1:
        test_DEM_np_new_burned_wide = np.copy(test_DEM_np_new_burned)

    # loop over all specified upstream points
    for k in range(len(list_upstream_point_locs)):
        
        # value to control while loop
        found_pit = 0
        
        # starting location
        last_location  = list_upstream_point_locs[k]
        
        # write values at starting location to lists
        if p == 0:
            DEM_list_original[k]    = [test_DEM_np[last_location]]
            DEM_list_adjusted[k]    = [test_DEM_np_new[last_location]]
            DEM_list_burned_main[k] = [test_DEM_np_new_burned[last_location]]
            
            list_locations_loop[k]  = [last_location]
        
        # loop from upstream starting location to most downstream point
        while found_pit == 0:
            
            # direction at last known location
            temp_direction = test_LDD_np[last_location]
            
            # find new location based on direction and LDD type
            if LDD_type == 'PCRaster':
                if temp_direction == 5:
                    found_pit = 1
                    break
                elif temp_direction == 6:
                    new_location = (last_location[0] + 0, last_location[1] + 1)
                elif temp_direction == 3:
                    new_location = (last_location[0] + 1, last_location[1] + 1)
                elif temp_direction == 2:
                    new_location = (last_location[0] + 1, last_location[1] + 0)
                elif temp_direction == 1:
                    new_location = (last_location[0] + 1, last_location[1] - 1)
                elif temp_direction == 4:
                    new_location = (last_location[0] + 0, last_location[1] - 1)
                elif temp_direction == 7:
                    new_location = (last_location[0] - 1, last_location[1] - 1)
                elif temp_direction == 8:
                    new_location = (last_location[0] - 1, last_location[1] + 0)
                elif temp_direction == 9:
                    new_location = (last_location[0] - 1, last_location[1] + 1)
            elif LDD_type == 'HydroSHEDS':
                if temp_direction == 0:
                    found_pit = 1
                    break
                elif temp_direction == 1:
                    new_location = (last_location[0] + 0, last_location[1] + 1)
                elif temp_direction == 2:
                    new_location = (last_location[0] + 1, last_location[1] + 1)
                elif temp_direction == 4:
                    new_location = (last_location[0] + 1, last_location[1] + 0)
                elif temp_direction == 8:
                    new_location = (last_location[0] + 1, last_location[1] - 1)
                elif temp_direction == 16:
                    new_location = (last_location[0] + 0, last_location[1] - 1)
                elif temp_direction == 32:
                    new_location = (last_location[0] - 1, last_location[1] - 1)
                elif temp_direction == 64:
                    new_location = (last_location[0] - 1, last_location[1] + 0)
                elif temp_direction == 128:
                    new_location = (last_location[0] - 1, last_location[1] + 1)
                    
            # check if while-loop should stop here because value was already looped over before
            if k > 0:
                if stop_same_point == 1:
                    if new_location in list_locations_loop[0]:
                        # write this final location to list of locations
                        list_locations_loop[k].append(new_location)
                        # stop while loop
                        found_pit = 1
                        break
            
            # if first loop, apply smoothing algorithm
            if p == 0:
                # write new location to list of locations
                list_locations_loop[k].append(new_location)
                
                # compare DEM value at new location with last location
                temp_difference_DEM = test_DEM_np[new_location] - test_DEM_np_new[last_location]
                
                # adjust value if DEM new location > DEM last location
                if temp_difference_DEM > 0.:
                    test_DEM_np_new[new_location] = test_DEM_np_new[last_location]
                    
                # if DEM new location < DEM last location, assess whether this is an outlier by checking DEM values at the next 
                # locations.
                # for efficiency, it can only be an outlier if the difference is larger/equal than a certain threshold value
                elif temp_difference_DEM <= (outlier_threshold * -1):
                    
                    temp_list_locations = []
                    temp_list_locations.append(new_location)
                    
                    for i in range(outlier_points):
                        
                        # direction at new location
                        temp_new_direction = test_LDD_np[new_location]
                        
                        # find next location based on direction and LDD type
                        if LDD_type == 'PCRaster':
                            if temp_new_direction == 5:
                                found_pit = 1
                                break
                            elif temp_new_direction == 6:
                                temp_list_locations.append((temp_list_locations[i][0] + 0, temp_list_locations[i][1] + 1))
                            elif temp_new_direction == 3:
                                temp_list_locations.append((temp_list_locations[i][0] + 1, temp_list_locations[i][1] + 1))
                            elif temp_new_direction == 2:
                                temp_list_locations.append((temp_list_locations[i][0] + 1, temp_list_locations[i][1] + 0))
                            elif temp_new_direction == 1:
                                temp_list_locations.append((temp_list_locations[i][0] + 1, temp_list_locations[i][1] - 1))
                            elif temp_new_direction == 4:
                                temp_list_locations.append((temp_list_locations[i][0] + 0, temp_list_locations[i][1] - 1))
                            elif temp_new_direction == 7:
                                temp_list_locations.append((temp_list_locations[i][0] - 1, temp_list_locations[i][1] - 1))
                            elif temp_new_direction == 8:
                                temp_list_locations.append((temp_list_locations[i][0] - 1, temp_list_locations[i][1] + 0))
                            elif temp_new_direction == 9:
                                temp_list_locations.append((temp_list_locations[i][0] - 1, temp_list_locations[i][1] + 1))
                        elif LDD_type == 'HydroSHEDS':
                            if temp_new_direction == 0:
                                found_pit = 1
                                break
                            elif temp_new_direction == 1:
                                temp_list_locations.append((temp_list_locations[i][0] + 0, temp_list_locations[i][1] + 1))
                            elif temp_new_direction == 2:
                                temp_list_locations.append((temp_list_locations[i][0] + 1, temp_list_locations[i][1] + 1))
                            elif temp_new_direction == 4:
                                temp_list_locations.append((temp_list_locations[i][0] + 1, temp_list_locations[i][1] + 0))
                            elif temp_new_direction == 8:
                                temp_list_locations.append((temp_list_locations[i][0] + 1, temp_list_locations[i][1] - 1))
                            elif temp_new_direction == 16:
                                temp_list_locations.append((temp_list_locations[i][0] + 0, temp_list_locations[i][1] - 1))
                            elif temp_new_direction == 32:
                                temp_list_locations.append((temp_list_locations[i][0] - 1, temp_list_locations[i][1] - 1))
                            elif temp_new_direction == 64:
                                temp_list_locations.append((temp_list_locations[i][0] - 1, temp_list_locations[i][1] + 0))
                            elif temp_new_direction == 128:
                                temp_list_locations.append((temp_list_locations[i][0] - 1, temp_list_locations[i][1] + 1))
                                
                    # find DEM values at these locations and difference with value at last location
                    temp_new_DEM_values     = []
                    temp_new_difference_DEM = []
                    for i in range(1,len(temp_list_locations)):
                        temp_new_DEM_values.append(test_DEM_np_new[temp_list_locations[i]])
                        temp_new_difference_DEM.append(temp_new_DEM_values[i-1] - test_DEM_np_new[last_location])
                        
                    # if the mean difference of these points is less (less negative=larger) than the difference with only the new 
                    # location, declare the new location an outlier and assign it a value equal to the average of the next values, 
                    # but no larger than the last value
                    if np.mean(temp_new_difference_DEM) > temp_difference_DEM:
                        # prints checking the rough location of the inner delta where I had the issue of too much burning before:
                        # commented out for now, since it seems to be ok now and I can just check the graphs
                        #if ((count+1)>=1500) and ((count+1)<=1600):
                        #    print count+1
                        #    print test_DEM_np_new[last_location]
                        #    print test_DEM_np[new_location]
                        #    print temp_new_DEM_values
                        #    print min(np.ceil(np.mean(temp_new_DEM_values)), test_DEM_np_new[last_location])
                        test_DEM_np_new[new_location] = min(np.ceil(np.mean(temp_new_DEM_values)), test_DEM_np_new[last_location])
                        
                    # and if this is not the case, it is not an outlier and the code will not making any changes
                
                # apply main stream burning
                test_DEM_np_new_burned[new_location] = test_DEM_np_new[new_location]
                test_DEM_np_new_burned[new_location] -= apply_burning_main
                if limit_to_zero == 1:
                    test_DEM_np_new_burned[new_location] = max(0., test_DEM_np_new_burned[new_location])
                
                # add values to lists
                DEM_list_original[k].append(test_DEM_np[new_location])
                DEM_list_adjusted[k].append(test_DEM_np_new[new_location])
                DEM_list_burned_main[k].append(test_DEM_np_new_burned[new_location])
                    
                # update location
                last_location = new_location
                
                # count iterations
                count[k] += 1
            
            # if not first loop, apply burning next to main stream
            # (burn everywhere next to new location, except at locations that were already burned)
            elif p == 1:
                
                # find to-be-burned locations
                temp_to_be_burned_locations = []
                temp_to_be_burned_locations.append((new_location[0] + 0, new_location[1] + 1))
                temp_to_be_burned_locations.append((new_location[0] + 1, new_location[1] + 1))
                temp_to_be_burned_locations.append((new_location[0] + 1, new_location[1] + 0))
                temp_to_be_burned_locations.append((new_location[0] + 1, new_location[1] - 1))
                temp_to_be_burned_locations.append((new_location[0] + 0, new_location[1] - 1))
                temp_to_be_burned_locations.append((new_location[0] - 1, new_location[1] - 1))
                temp_to_be_burned_locations.append((new_location[0] - 1, new_location[1] + 0))
                temp_to_be_burned_locations.append((new_location[0] - 1, new_location[1] + 1))
                
                # find any locations that have already been looped over or that have a missing value in the landmask
                temp_indices_to_remove = []
                for q in range(len(temp_to_be_burned_locations)):
                    if (temp_to_be_burned_locations[q] in list_locations_loop[0]) or \
                       (temp_to_be_burned_locations[q] in list_locations_loop[1]) or \
                       (temp_to_be_burned_locations[q] in list_locations_loop_burn_wide) or \
                       (test_landmask_np[temp_to_be_burned_locations[q]] == 255):
                        temp_indices_to_remove.append(q)
                
                # remove these from list (use reversed order, going from high to low index to keep the correct indices order)
                if len(temp_indices_to_remove) > 0:
                    for q in sorted(temp_indices_to_remove, reverse=True):
                        del temp_to_be_burned_locations[q]
                        
                # add final to-be-burned locations to list
                if len(temp_to_be_burned_locations) > 0:
                    for q in range(len(temp_to_be_burned_locations)):
                        list_locations_loop_burn_wide.append(temp_to_be_burned_locations[q])
                    
                # burn these locations
                if len(temp_to_be_burned_locations) > 0:
                    for q in range(len(temp_to_be_burned_locations)):
                        test_DEM_np_new_burned_wide[temp_to_be_burned_locations[q]] -= apply_burning_wide
                        if limit_to_zero == 1:
                            test_DEM_np_new_burned_wide[temp_to_be_burned_locations[q]] = \
                                                                max(0., test_DEM_np_new_burned_wide[temp_to_be_burned_locations[q]])
                        
                # update location
                last_location = new_location
```
